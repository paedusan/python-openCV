# Gradients

[from https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_gradients/py_gradients.html]

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pylab # this allows you to control figure size 
pylab.rcParams['figure.figsize'] = (10.0, 8.0) # this controls figure size in the notebook
```

```python
img = cv2.imread('data/sudoku2.jfif',0)
```

```python
plt.imshow(img,'gray')
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
plt.imshow(sobelx,'gray')
```

```python
plt.imshow(sobely,'gray')
```

```python
plt.imshow(laplacian,'gray')
```

* Hmmm, not so good... what if?

```python
ret,th1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
plt.imshow(th1,'gray')
```

* Change the threshold!

```python
ret,th = cv2.threshold(img,150,255,cv2.THRESH_BINARY)
plt.imshow(th,'gray')
```

* Let's see what Laplacian looks like!

```python
sobelx2 = cv2.Sobel(th,cv2.CV_64F,1,0,ksize=5)
sobely2 = cv2.Sobel(th,cv2.CV_64F,0,1,ksize=5)
laplacian2 = cv2.Laplacian(th,cv2.CV_64F)
```

```python
plt.imshow(laplacian2,'gray')
```

```python
plt.imshow(sobelx2,'gray')
```

```python
plt.imshow(sobely2,'gray')
```

### Blend!

```python
blended = cv2.addWeighted(src1=sobelx2,alpha=0.5,src2=sobely2,beta=0.5,gamma=0)
```

```python
plt.imshow(blended,'gray')
```
