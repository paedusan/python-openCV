# Opening Image Files using OpenCV

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('data/puppy2.jfif')
```

```python
img
```

------------

```python
img_bgr = cv2.imread('data/puppy2.jfif')
plt.imshow(img_bgr)
```

A Vulcan dog?? No, the default of openCV is BGR, but Matplotlib expects an RGB so we can use openCV conversion functions cv2.cvtColor() (if you don't want to work directly with the numpy array).

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('data/puppy2.jfif',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray = cv2.imread('data/puppy2.jfif',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

## Resize Images

```python
img_rgb.shape
# width, height, color channels
```

```python
img =cv2.resize(img_rgb,(1300,275))
```

```python
plt.imshow(img)
```

### By ratio

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
%matplotlib qt
plt.imshow(new_img)
```

### Flipping Images

```python
%matplotlib inline
```

```python
# Along central x axis
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
# Along both axis
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```
